# EDA - univariate , bivariate , multivariate 

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
print(tips)
```

```python
tips.head()
```

```python
## scatter plot for numerical and numerical data analysis 
sns.scatterplot(
    x = 'total_bill',
    y = 'tip',
    data = tips,
    hue = 'sex',
    size = 'size',
    style = "smoker"
    
)
```

```python
# barplot - numerical and categorical data
sns.barplot(
    x = 'sex',
    y = 'tip',
    data = tips,
    hue = 'smoker'
)
# shows the tips data between smoker men vs nonsmoker men and smoker women and non smoker women  
```

```python
# box plot - numerical and categorical dataa 
sns.boxplot(
    x = 'sex',
    y = 'tip',
    data = tips 
    
) # shows us the outliers 
```

```python
# displot ( numerical - categorical data )
sns.displot(
    x = 'sex',
    y = 'tip',
    data = tips 
)
```

```python
# heatmap (categorical - categorical)
sns.heatmap(pd.crosstab(
    tips['sex'],
    tips['smoker']
))
```

```python
# clustormap = categorical - categorical 
sns.clustermap(pd.crosstab(
    tips['sex'],
    tips['smoker'],
    
))
```

```python
#pairplot(multiple scatterplot)
sns.pairplot(tips)
```
